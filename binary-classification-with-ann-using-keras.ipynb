```python
#Loading the libraries / methods:

import numpy as np
import pandas as pd
import keras
from sklearn.model_selection  import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import confusion_matrix, accuracy_score
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
#Loading the datasets:

pred_breast = pd.read_csv("/kaggle/input/breastcancerwisconsin/features_breast.csv")
class_breast = pd.read_csv("/kaggle/input/breastcancerwisconsin/class_breast.csv")
```

```python
print(pred_breast.shape, class_breast.shape)
```

```python
pred_breast.head()
```

```python
#Builnding a neural network with two hidden layer:

def NeuralNet():
    classifier = Sequential()
    classifier.add(Dense(units = 16, activation='relu', kernel_initializer='random_uniform', input_dim=30))
    classifier.add(Dropout(0.2))
    classifier.add(Dense(units = 16, activation='relu', kernel_initializer='random_uniform'))
    classifier.add(Dropout(0.2))  
    classifier.add(Dense(units = 1, activation='sigmoid')) #output layer

    #Setting the optimizer, loss function, and metrics:
    opt = keras.optimizers.Adam(lr = 0.001, decay = 0.0001, clipvalue=0.5)
    classifier.compile(optimizer=opt, loss='binary_crossentropy', metrics=['binary_accuracy'])
    
    return classifier
```

```python
#Hyperparameter optimization (tuning):

'''
parameters = {'batch_size': [10, 30],
              'epochs': [50, 100],
              'optimizer': ['adam', 'sgd'],
              'loss': ['binary_cross_entropy', 'hinge']
              'kernel_initializer': ['random_uniform', 'normal'], 
              'activation': ['relu', 'tanh'],
              'neurons': [16, 8]}

grid_search = GridSearchCV(estimator = classifier, param_grid = parameters, scoring = 'accuracy', cv = 5)
best_param = grid_search.best_params_ 
best_acc = grid_search.best_score_
'''
```

```python
classifier = KerasClassifier(build_fn = NeuralNet, epochs=100, batch_size=10)
```

```python
#10-fold cross-validation:

results = cross_val_score(estimator = classifier, X = pred_breast, y = class_breast, cv = 10, scoring = 'accuracy')
results
```

```python
mean = results.mean()
std = results.std()
print(mean)
print(std)
```
